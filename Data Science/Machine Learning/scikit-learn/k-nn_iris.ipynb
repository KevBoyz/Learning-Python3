```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()

iris.data  # Features X
iris.target  # Labels y

iris.target_names
```

```python
# Transformando em df pandas

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target number'] = iris.target

# Adicionando coluna de labels com nome
iris_df['target names'] = pd.Categorical.from_codes(iris.target, iris.target_names)

iris_df.head()
```

```python
%matplotlib inline 
```

```python
iris_df.plot.scatter('sepal length (cm)', 'sepal width (cm)', c='target number', cmap='Set1')
```

```python
# Dividir teste e treinamento

X, y = load_iris(return_X_y=True)  # Modo oficial de importar 
X_train, y_train, X_test, y_test = train_test_split(X, y, test_size=0.33, random_state=42)  # Seed para fazer radomização
```
