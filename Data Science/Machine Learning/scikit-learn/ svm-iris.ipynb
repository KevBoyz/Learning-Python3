```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import svm
import pandas as pd
import seaborn as sns
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

%matplotlib inline
iris_df.plot.scatter('sepal length (cm)', 'sepal width (cm)', c='target', cmap='Set1')
```

O modelo svm cria uma reta entre as classes. A reta é o mais equidistante possível dos pontos de ambos os lados. Isso para o modelo de  
kernel linear, existem diferentes kernels como polinomial ou rbf.

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

clf = svm.SVC(C=1.0)
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```
